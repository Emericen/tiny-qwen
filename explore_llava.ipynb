# LLaVA Dataset Exploration

```python
from data.llava import LLaVADataset
import matplotlib.pyplot as plt
```

```python
# Load the dataset (will download if needed)
dataset = LLaVADataset()
```

```python
# Show dataset info
print(f"Dataset size: {len(dataset)}")
print(f"First item keys: {list(dataset[0].keys())}")
```

```python
# Get first few samples
sample = dataset[0]

print(f"ID: {sample['id']}")
print(f"Messages: {sample['messages']}")
print(f"Image size: {sample['image'].size}")

# Display the image
plt.figure(figsize=(6, 6))
plt.imshow(sample['image'])
plt.title(f"Sample: {sample['id']}")
plt.axis('off')
plt.show()
```

```python
# Show a few more examples
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i in range(6):
    sample = dataset[i]
    axes[i].imshow(sample['image'])
    axes[i].set_title(f"ID: {sample['id']}\nResponse: {sample['messages'][1]['content'][:50]}...", fontsize=8)
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```
